```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import scipy as sp
from scipy import signal
from datetime import timedelta, datetime

'''
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
'''
```

```python
NT = 20000
dt = 1/10
f_N = 1/(2*dt)

A = [250, 34, 8, 10]

w1 = 2*np.pi*f_N/100
w2 = 4.3*w1
w3 = 4.7*w1
w4 = 2*np.pi*f_N/500

W = [w1, w2, w3, w4]
```

```python
def CN(sigma):
    return sigma*np.random.randn(NT)

def signal(NT, dt, A, W):
    sig = np.zeros(NT)
    t = np.arange(0, NT, 1)
    
    for a,w in zip(A,W):
        sig += a*np.cos(w*t)
    
    return sig
```

```python
for a, w in zip(A,W):
    print(a)
    print(w)
```

```python
x1 = CN(20)+signal(NT, dt, A, W)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x1)
ax.set_xlim(0,100)
```

```python
X1 = np.fft.fft(x1)
ii2 = range(0,int(NT/2))

Gxx = X1[ii2]*np.conj(X1[ii2])/(NT/2)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.loglog(Gxx)
```
