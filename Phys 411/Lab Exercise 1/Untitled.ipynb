```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import scipy as sp
from scipy import signal
from datetime import timedelta, datetime
```

The parameters below are defined such that there are $NT=20000$ data points with a samply frequency of $f=1/dt=10~\text{Hz}$. The Nyquist frequency (lowest possible resolved frequency) is given by $f_N=f/2$. The amplitudes and frequencies of the signal are chosen to match what is required on the assignment.

```python
NT1 = 20000
NT2 = 2000
dt = 1/10
f_N = 1/(2*dt)

A = [250, 34, 8, 10]

w1 = 2*np.pi*f_N/100
w2 = 4.3*w1
w3 = 4.7*w1
w4 = 2*np.pi*f_N/500

W = [w1, w2, w3, w4]
```

The functions below are used to create the signal and the noise.

```python
def CN(sigma, N):
    return sigma*np.random.randn(N)

def signal(NT, dt, A, W):
    sig = np.zeros(NT)
    t = np.arange(0, NT, 1)*dt
    
    for a,w in zip(A,W):
        sig += a*np.cos(w*t)
    
    return sig

# Signal created ...
t1 = np.arange(0, NT1, 1)*dt
x1 = CN(20, NT1)+signal(NT1, dt, A, W)
t2 = np.arange(0, NT2, 1)*dt
x2 = CN(20, NT2)+signal(NT2, dt, A, W)
```

## Plot The Signal

For dramatic effect, the bottom plot's axis has been scaled so we can see how the signals differ in time.

```python
fig, axes = plt.subplots(2, 1, figsize=(16,10))

axes[0].plot(t1,x1)
axes[0].set_title('Signal Sample 1')
axes[0].set_ylabel('Amplitude [arb]')

axes[1].plot(t2,x2)
axes[1].set_title('Signal Sample 2')
axes[1].set_xlabel('Time (seconds)')
axes[1].set_ylabel('Amplitude [arb]')
axes[1].set_xlim(axes[0].get_xlim())

plt.show()
```

Below we take the Fourier transform of the signal and use it to plot the power spectral density $G_{xx}(f)$. We only include positive frequencies.

```python
X1 = np.fft.fft(x1)
X2 = np.fft.fft(x2)
ii1 = range(0,int(NT1/2))
ii2 = range(0,int(NT2/2))

# Signal 1
fq1 = np.arange(0, int(NT1/2), 1)/(NT1 * dt)
Gxx1 = np.abs(X1[ii1])**2 /(NT1/2)

# Signal 2
fq2 = np.arange(0, int(NT2/2), 1)/(NT2 * dt)
Gxx2 = np.abs(X2[ii2])**2 /(NT2/2)
```

## Plot the 1-Sided Power Spectral Density

```python
fig, axes = plt.subplots(2, 1, figsize=(12,10))
axes[0].loglog(fq1,Gxx1)
axes[0].set_title('1-Sided Power Spectral Density (Signal 1)')
axes[0].set_ylabel('$G_{xx}[V^2/Hz]$')

axes[1].loglog(fq2,Gxx2)
axes[1].set_title('1-Sided Power Spectral Density (Signal 2)')
axes[1].set_xlabel('Frequency [Hz]')
axes[1].set_ylabel('$G_{xx}[V^2/Hz]$')

plt.show()
```

The signal which was measured for less time predicts more energy in low frequencies and less energy in high frequencies. For Signal 2, we can't determine the peaks at higher frequencies.
