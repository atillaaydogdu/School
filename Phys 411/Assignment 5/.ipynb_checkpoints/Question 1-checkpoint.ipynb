# Getting the Data

```python
def matlab_to_python_datetime(matlab_datenum):
    return datetime.fromordinal(int(matlab_datenum)) + timedelta(days=matlab_datenum%1) - timedelta(days = 366)
```

## Importing Minute Resolution Data

Now we import the minute resolution data, which only contains data from the one UvicSci temperature data station. We begin by getting the time information, which contains the the start and end time, and the number of measurements taken.

```python
time_info = np.genfromtxt('JamesBay_temperature.dat', delimiter=" ", max_rows=3)
t0 = time_info[0]
tf = time_info[1]
N = time_info[2]
```

Since the time between measurements is always one minute, we can generate an array of times of measurements based on the information above.

```python
incrementer = (tf-t0)/N
timearray = np.arange(t0, tf, incrementer)
```

We now open  the actual temperature data, and create a dataframe using the appropriate timestamps.

```python
df_james = pd.read_csv('JamesBay_temperature.dat', sep=" ", skiprows=[0,1,2], names=['Temperature'])
df_james.index = [matlab_to_python_datetime(i) for i in timearray]
df_james.index = [i - timedelta(hours = 7) for i in df_james.index] #Convert to PST
```

Same thing for Deep Cove.

```python
time_info = np.genfromtxt('DeepCove_temperature.dat', delimiter=" ", max_rows=3)
t0 = time_info[0]
tf = time_info[1]
N = time_info[2]
```

```python
incrementer = (tf-t0)/N
timearray = np.arange(t0, tf, incrementer)
```

```python
df_deep = pd.read_csv('DeepCove_temperature.dat', sep=" ", skiprows=[0,1,2], names=['Temperature'])
df_deep.index = [matlab_to_python_datetime(i) for i in timearray]
df_deep.index = [i - timedelta(hours = 7) for i in df_deep.index] #Convert to PST
```

The only data set with a significant amount of missing data is James Bay during the summer. We need to concatenate the data set to ensure that we only perform analysis on the first region.

Little trick to get consecutive NaN's in a row. See https://stackoverflow.com/questions/29007830/identifying-consecutive-nans-with-pandas

We look for more than 1000 NaN values in a row and concatenate the data accordingly.

```python
nulls = data_james1['Temperature'].isnull().astype(int).groupby(data_james1['Temperature'].notnull().astype(int).cumsum()).sum()
first_point_in_gap = nulls[nulls>1000].index[0]

data_james1_rem = data_james1[first_point_in_gap::] #first save the data that was removed
data_james1 = data_james1[1:first_point_in_gap] #then modify the original DataFrame
```

Convert the data into numpy arrays.

```python
data_james1 = data_james1.fillna(method='ffill')
data_james2 = data_james2.fillna(method='ffill')
james1 = np.array(data_james1['Temperature']); t_james1 = np.array(data_james1.index)
james2 = np.array(data_james2['Temperature']); t_james2 = np.array(data_james2.index); 
```

```python
data_deep1 = data_deep1.fillna(method='ffill')
data_deep2 = data_deep2.fillna(method='ffill')
deep1 = np.array(data_deep1['Temperature']); t_deep1 = np.array(data_deep1.index)
deep2 = np.array(data_deep2['Temperature']); t_deep2 = np.array(data_deep2.index); 
```

# Design the Filter

```python
def ellip_bandpass(lowcut, highcut, fs, rp, rs, order=5, btype='bandpass'):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    if (btype == 'bandpass'):
        b, a = sp.signal.ellip(order, rp, rs, [low, high], btype=btype)
    if (btype == 'lowpass'):
        b, a = sp.signal.ellip(order, rp, rs, low, btype=btype)
    if (btype == 'highpass'):
        b, a = sp.signal.ellip(order, rp, rs, high, btype=btype)
        
    return b, a

def ellip_bandpass_filter(data, lowcut, highcut, fs, rp, rs, order=5):
    
    #filter high signals
    b, a = ellip_bandpass(lowcut, highcut, fs, rp, rs, order=order, btype='lowpass')
    w_high, h_high = signal.freqz(b, a, 100000)
    y = signal.filtfilt(b, a, data)
    
    #filter low signals
    b, a = ellip_bandpass(lowcut, highcut, fs, rp, rs, order=order, btype='highpass')
    w_low, h_low = signal.freqz(b, a, 100000)
    y = signal.filtfilt(b, a, y)
    
    # Now normalize the strength of the filter
    scale = 1/np.max(abs(h_low*h_high))
    
    return y*scale
```

# Part 1

Get the properties of our filter, and filter for the diurnal band.

```python
b, a = ellip_bandpass(1/(25*hour), 1/(23*hour), (1/mint), 2, 50, order=4, btype='highpass')
w_high, h_high = signal.freqz(b, a, 100000)

b, a = ellip_bandpass(1/(25*hour), 1/(23*hour), (1/mint), 2, 50, order=4, btype='lowpass')
w_low, h_low = signal.freqz(b, a, 100000)

scale = 1/np.max(abs(h_low*h_high))
```

Plot frequency response by scaling by the nyquist frequency. The various curves on the plot are explained below:

```python
fs=0.5*(1/mint)

fig, ax = plt.subplots(figsize=(8,4))

ax.plot((fs / np.pi) * w_low, abs(h_low), label='Lowpass Component')
ax.plot((fs / np.pi) * w_high, abs(h_high), label='Highpass Component')
ax.plot((fs / np.pi) * w_low, abs(h_low*h_high), label='Bandpass')
ax.plot((fs / np.pi) * w_low, scale*abs(h_low*h_high), label='Filter Used', lw=5)

ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Filter Response [arb]')
ax.set_title('Frequency Response of Elliptical Filter')

ax.set_xlim(1/(40*hour), (1/(15*hour)))
ax.axvline(1/(24*hour), color='k', linestyle='--', label=('Diurnal Frequency'))


ax.grid()
ax.legend()

plt.show()
```

The lowpass and highpass components (blue and orange lines) are used to contruct the bandpass (green line) component. Since the filter response is less than 1, we need to rescale it; we do this by normalizing it by:

```python
np.max(abs(h_low*h_high))
```

We then get the red curve, which is the true filter used.

## Part C

The frequency in cycles per day is given by

$$ fs = \frac{1}{60\text{s}} \cdot \frac{24 \cdot 3600\text{s}}{\text{day}}=\frac{1440}{\text{day}} $$

```python
fs = 1440
```

Unfiltered Time Series:

```python
fj1, Xj1 = sp.signal.periodogram(james1, fs=fs)
fj2, Xj2 = sp.signal.periodogram(james2, fs=fs)
fd1, Xd1 = sp.signal.periodogram(deep1, fs=fs)
fd2, Xd2 = sp.signal.periodogram(deep1, fs=fs)
```

Diurnal Band:

```python
fj1_f, Xj1_f = sp.signal.periodogram(james1_filt, fs=fs)
fj2_f, Xj2_f = sp.signal.periodogram(james2_filt, fs=fs)
fd1_f, Xd1_f = sp.signal.periodogram(deep1_filt, fs=fs)
fd2_f, Xd2_f = sp.signal.periodogram(deep2_filt, fs=fs)
```

### Explanation

Clearly the percentage of diurnal variance is greater in the summer than in the winter (~4x bigger). This implies that the flucuation in temperature at a daily frequency is more prominent in the summer than in the winter. Perhaps the most logical explanation for this is the angle of elevation of the sun in the summer vs. winter. In the winter, the days are shorter and the sub spends more time in the horizon. In the summer, however, the sun comes directly overhead and then vanishes at night time. The variation in solar flux is thus stronger in the summer than in the winter, and this might account for the greater variance in the summer diurnal frequency.

Another factor that influences diurnal temperature fluctuations is the humidity in the air. In deserts, there is often extreme variations in temperature from day to night (Snake River Plain can have high daily temperature os 38$^\circ C$ and low nightly temperatures of $5-10 ^\circ C$). Places with higher humidity tend to have more water molecules in the air; which store energy (heat) over cold nightly periods. This results in less flucuations, and thus a lower diurnal frequency percentage. Perhaps an interesting extension to this assignment in the future could involve comparing the percentage of variance in the diurnal band between a coastal area (Victoria) and a desert area (Kelowna?).
