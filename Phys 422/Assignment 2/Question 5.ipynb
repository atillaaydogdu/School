```python
import numpy as np
import scipy.integrate as integrate
from scipy.integrate import quad, dblquad
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as colors

cmap = cm.Blues
```

The magnetic field of a line current is

$$\vec{B}=\frac{\mu_0 I}{4\pi}\int \frac{d\vec{l}' \times (\vec{r}-\vec{r}')}{|\vec{r}-\vec{r}'|^3} $$

In our situation we have

$$\vec{r}=z\hat{z}+x\hat{x} $$
$$\vec{r}'=R(\cos\phi \hat{x}+\sin\phi \hat{y}) $$
$$d\vec{l}'=Rd\phi \hat{\phi} = Rd\phi(-\sin\phi \hat{x}+ \cos\phi \hat{y}) $$

Plugging in all cross products yields

$$\vec{B}(x,z)=\frac{\mu_0 I}{4\pi} \int_0^{2\pi} \frac{[Rz\cos\phi] d\phi\hat{x}+[Rz\sin\phi] d\phi\hat{y}+[R^2-xR\cos\phi]d\phi \hat{z}}{(R^2-2xR\cos\phi+x^2+z^2)^{3/2}} $$

However since we don't have values for the curent, a more useful quantity to plot would be

$$\frac{4\pi}{\mu_0 I} \vec{B}(x,z)=\int_0^{2\pi} \frac{[Rz\cos\phi] d\phi\hat{x}+[Rz\sin\phi] d\phi\hat{y}+[R^2-xR\cos\phi]d\phi \hat{z}}{(R^2-2xR\cos\phi+x^2+z^2)^{3/2}} $$


```python
R=0.1
```

Define the integrand

```python
def integrand_x(phi, x,z):
    return (R*z*np.cos(phi))/(R**2-2*x*R*np.cos(phi)+x**2+z**2)**(3/2)

def integrand_y(phi, x,z):
    return (R*z*np.sin(phi))/(R**2-2*x*R*np.cos(phi)+x**2+z**2)**(3/2)

def integrand_z(phi, x,z):
    return (R**2-x*R*np.cos(phi))/(R**2-2*x*R*np.cos(phi)+x**2+z**2)**(3/2)
```

Define functions to get the magnitude and direction of the magnetic field.

```python
def mag_field(x,z):
    x_comp = integrate.quad(integrand_x, 0, 2*np.pi, args=(x,z))[0]
    z_comp = integrate.quad(integrand_z, 0, 2*np.pi, args=(x,z))[0]
    return np.array([x_comp, z_comp])

def mag_magnitude(x,z):
     return np.sqrt(mag_field(x,z)[0]**2+mag_field(x,z)[1]**2)
    
def mag_dir(x,z):
    return mag_field(x,z)/mag_magnitude(x,z)
```

Hackish way to get all points.

```python
X_d, Z_d = np.meshgrid(np.linspace(-4*R, 4*R, 17), np.linspace(-4*R, 4*R, 17))
gridtuples = np.dstack((X_d, Z_d))

U_d = np.empty(X_d.shape)
V_d = np.empty(X_d.shape)
for i, row in enumerate(gridtuples):
    for j, ele in enumerate(gridtuples[i]):
        U_d[i,j]=mag_dir(ele[0], ele[1])[0]
        V_d[i,j]=mag_dir(ele[0], ele[1])[1]

X_m, Z_m = np.meshgrid(np.linspace(-4*R, 4*R, 177), np.linspace(-4*R, 4*R, 177))
gridtuples = np.dstack((X_m, Z_m))        

mag = np.empty(X_m.shape)
for i, row in enumerate(gridtuples):
    for j, ele in enumerate(gridtuples[i]):
        mag[i,j]=mag_magnitude(ele[0], ele[1])
```

Do the plot.

```python
fig, ax = plt.subplots(figsize=(10,8))

cs1 = ax.pcolor(X_m, Z_m, mag, cmap=cmap, norm=colors.LogNorm(vmin=0.1, vmax=100))
cbar = fig.colorbar(cs1, ax=ax, extend='both', label=r'$\frac{4\pi}{\mu_0 I}\vec{B}$')

Q = ax.quiver(X_d, Z_d, U_d, V_d, units='width', width=0.002)

ax.scatter(np.array([-0.1, 0.1]), np.array([0,0]), color='r', label='Ring Location')

ax.set_title('Cross Section of Magnetic Field From Ring')
ax.set_xlabel('$x [cm]$')
ax.set_ylabel('$z [cm]$')

plt.show()
```

## Part B

We can parameterize the rotating loop of wire by

$$\vec{r}(u,v)=u\hat{x}+v\cos(\omega t)\hat{y}+(Z+v\sin(\omega t))\hat{z} $$

The flux is then given by $\vec{B} \times (\vec{r}_u \times \vec{r}_v)$ or

$$\text{Flux} = \sin(wt)\vec{B} \cdot \hat{y}+\cos(\omega t) \vec{B} \cdot \vec{z} $$

More precisely, the integrand $\vec{B} \cdot d\vec{A}$ is given by

$$\vec{B} \cdot d\vec{A} = \frac{\mu_0 I \sin(\omega t)}{4\pi} \int_0^{2\pi}\frac{Rz\sin\phi d\phi}{(R^2-2\sqrt{u^2 +v^2\cos^2(\omega t)}R\cos\phi+u^2+v^2 \cos^2(\omega t)+z^2)^{3/2}}  +\frac{\mu_0 I \cos(\omega t)}{4\pi} \int_0^{2\pi}\frac{R^2-\sqrt{u^2 +v^2\cos^2(\omega t)}R\cos\phi d\phi}{(R^2-2\sqrt{u^2 +v^2\cos^2(\omega t)}R\cos\phi+u^2+v^2 \cos^2(\omega t)+z^2)^{3/2}}  $$

and thus the total flux is given by

$$\text{Flux} = \iint_{u^2+v^2<1} \vec{B} \cdot d\vec{A} $$

```python
def flux(u, v, t):
    return np.sin(t)*mag_field(np.sqrt(u**2+(np.cos(t)*v)**2), Z+v*np.sin(t))[0] \
            + np.cos(t)*mag_field(np.sqrt(u**2+(np.cos(t)*v)**2), Z+v*np.sin(t))[1]
```

```python
Z = 0.12
t = 3*np.pi/2
r_lit = 0.05

du = r_lit/100
```

```python
bounds = np.arange(-r_lit, r_lit, du)
```

```python
def compute_flux(t):
    FLUX = 0
    for u in bounds:
        for v in bounds:
            if u**2+v**2<r_lit**2:
                FLUX += flux(u,v,t)*du**2
    return FLUX
```

```python
def compute_flux(t):
    return dblquad(lambda A, B: flux(A*np.cos(B), A*np.sin(B), t), 0, 2*np.pi, 0, r_lit)
```

```python
t = np.arange(0, 2*np.pi, 0.1)
f = np.array([compute_flux(i)[0] for i in t])

fig, ax = plt.subplots(figsize=(10,8))

ax.plot(t,f)
ax.plot(t,f.max()*np.cos(t))
```
