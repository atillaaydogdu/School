```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 1
q = 1
```

Takes in the charge $q$ and location of the charge $\vec{R}$ - returns the electric field at an arbitrary point $\vec{X}$.

```python
def elec_field(q, R, X):
    rx = R[0]; ry = R[1]
    x = X[0]; y=X[1]
    return (q/((x-rx)**2+(y-ry)**2)**(3/2))*np.array([x-rx, y-ry])
```

```python
def charge_arrangement(X):
    pnt_1, q1 = np.array([0, a]), q
    pnt_2, q2 = np.array([0, -a]), q
    pnt_3, q3 = np.array([a, 0]), -q
    pnt_4, q4 = np.array([-a, 0]), -q
    
    return (elec_field(q1, pnt_1, X) + elec_field(q2, pnt_2, X) + 
        elec_field(q3, pnt_3, X) + elec_field(q4, pnt_4, X) )
```

```python
def norm_charge_arrangement(X):
    return charge_arrangement(X)/np.linalg.norm(charge_arrangement(X))
```

```python
norm_charge_arrangement(np.array([5,5]))
```

Define a meshgrid

```python
X, Y = np.meshgrid(np.arange(-2*a, 2*a, .2), np.arange(-2*a, 2*a , .2))
U = norm_charge_arrangement([X[0][0], Y[0][0]])
V = norm_charge_arrangement(Y)
```

```python
U
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Arrows scale with plot width, not view')
Q = ax1.quiver(X, Y, U, V, units='width')
qk = ax1.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
```
