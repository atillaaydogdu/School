```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100,1)
plt.scatter(x, poisson.pmf(0.01, x), s=1)
```

```python
def poisson(n, v):
    return np.exp(-v)*(v**n)/np.math.factorial(n)
```

```python
poisson(10,5)
```

```python
def f(v):
    return poisson(0,v)+poisson(1,v)-0.9
```

```python
v = np.linspace(0.530,0.535,1000)
```

```python
plt.plot(v,f(v))
plt.grid()
```

```python
t = np.linspace(0,20,1000)
tau=t
```

```python
plt.figure(figsize=(6,4))
plt.plot(t,tau, label=r'$\hat{\tau}$')
plt.plot(t-5,tau, color='g', ls='--', label='90% lower bound for t')
plt.plot(t+5,tau, color='r', ls='--', label='90% upper bound for t')
plt.xlabel('t')
plt.ylabel(r'$\tau$')
plt.axhline(10, color='k', ls='--')
plt.scatter([15],[10], color='r')
plt.scatter([5],[10], color='g')
plt.xlim(0,20)
plt.legend()
plt.savefig('plots/chap7_ex2.png', dpi=300)
```
