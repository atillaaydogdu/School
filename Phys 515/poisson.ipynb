# Poisson confidence intevals

```python
import numpy as np
from scipy import stats

%matplotlib inline
import matplotlib.pyplot as plt
```

Change alpha and beta below to find the corresponding confidence interval on $\nu$ given $n$ observed events

```python
alpha=0.05; beta=0.05
lower=0; upper=0
lowlim=[0]; uplim=[]
for nu in np.arange(0.,25.,0.002):
    if (1-stats.poisson.cdf(lower,nu)>alpha):
        lowlim.append(nu)
        lower+=1
    if (stats.poisson.cdf(upper,nu)<beta):
        uplim.append(nu)
        upper+=1
print(" n  lower  upper")
for n in range(11):
    print("{0:2d} {1:5.2f}  {2:5.2f}".format(n,lowlim[n],uplim[n]))
```

```python
x = np.arange(0,10,0.1)

def f1(x):
    return x**2
def f2(x):
    return (x+0.5)**2
```

```python
plt.plot(x,f1(x), color='k', ls='--')
plt.plot(x,f2(x), color='k', ls='--')
plt.ylabel(r'$\theta$')
plt.xlabel('x')

plt.xlim(4,6)
plt.ylim(20,40)

plt.savefig('chap7_example.png', dpi=300)
```

```python
x = np.arange(0,4,0.01)
def f(x,tau):
    return 1/tau * np.e**(-x/tau)
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,f(x,1), color='b', label=r'Exponential $\tau=1$')
plt.plot(x,f(x,1.1), color='r', label=r'Exponential $\tau=1.1$')
plt.ylabel(r'$\tau$')
plt.xlabel('$x$')

plt.axvline(np.log(1/0.95), color='b', ls='--')
plt.axvline(np.log(1/0.05), color='b', ls='--')

plt.axvline(1.1*np.log(1/0.95), color='r', ls='--')
plt.axvline(1.1*np.log(1/0.05), color='r', ls='--')

plt.legend()

plt.savefig('plots/chap7_example2.png', dpi=300)
```
