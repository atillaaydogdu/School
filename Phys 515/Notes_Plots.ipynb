```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

from scipy import signal
```

# 5 Testing Hypotheses

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
```

```python
x = np.linspace(-30,30,1000)
y1 = gaussian(x, -7, 5)
y2 = gaussian(x, 7, 5)
```

```python
plt.figure(figsize=(7,2))
plt.plot(x,y1, label='$f(x|H_0)$')
plt.plot(x,y2, label='$f(x|H_1)$')
plt.axvline(3, color='k', ls='--')
plt.ylabel('Probability Density')
plt.xlabel('Data Values')
plt.legend()
plt.savefig('plots/chap5_plot1.png', dpi=300)
```

```python
x = np.linspace(-30,30,1000)
y1 = gaussian(x, 0, 5)
y2 = gaussian(x, 15, 5)
y3 = gaussian(x, -15, 5)
y4 = gaussian(x, 0, 1)

xfill1 = x[x>2*5]; yfill1 = y1[x>2*5]
xfill2 = x[x<-2*5]; yfill2 = y1[x<-2*5]
xfill3 = x[(x>-1) & (x<1)]; yfill3 = y1[(x>-1) & (x<1)]
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,2))
ax[0].plot(x,y1, label='$f(x|H_0)$')
ax[0].plot(x,y2, label='$f(x|H_1)$')
ax[0].fill_between(xfill1, 0, yfill1, facecolor='red', alpha=0.5)
ax[0].legend()

ax[1].plot(x,y1, label='$f(x|H_2)$')
ax[1].plot(x,y3, label='$f(x|H_3)$')
ax[1].fill_between(xfill2, 0, yfill2, facecolor='red', alpha=0.5)

ax[2].plot(x,y1, label='$f(x|H_2)$')
ax[2].plot(x,y4, label='$f(x|H_3)$')
ax[2].fill_between(xfill3, 0, yfill3, facecolor='red', alpha=0.5)
plt.savefig('plots/chap5_plot2.png', dpi=300)
```

```python
x = np.linspace(-30,30,1000)
y1 = gaussian(x, 0, 5)
y2 = gaussian(x, 15, 5)

y3 = gaussian(x, 0, 1)
y4 = gaussian(x, 15, 1)
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,2))
ax[0].plot(x,y1, label='$f(x|H_0)$')
ax[0].plot(x,y2, label='$f(x|H_1)$')
ax[0].set_title('pdfs for $X$')
ax[0].legend()

ax[1].plot(x,y3, label=r'$f_{\bar{X}}(x|H_0)$')
ax[1].plot(x,y4, label=r'$f_{\bar{X}}(x|H_1)$')
ax[1].set_title(r'pdfs for $\bar{X}$ with $n=10$')
ax[1].legend()

plt.savefig('plots/chap5_plot3.png', dpi=300)
```

```python
from scipy.stats import chi2

rv = chi2(4)
x = np.linspace(0,20,1000)
y = rv.pdf(x)

xfill1 = x[x>9.6]; yfill1 = y[x>9.6]; 
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,y)
plt.fill_between(xfill1, 0, yfill1, facecolor='red', alpha=0.5)
plt.axvline(9.6, color='k', ls='--')
plt.text(10, 0.05, '$p$={:.4f}'.format(1-rv.cdf(9.6)), fontsize=12) 
plt.xlabel('$\chi^2$')
plt.ylabel('Probability Density')
plt.savefig('plots/chap5_plot4.png', dpi=300)
```
