```python
% matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import cm
```

```python
j = complex(0,1)
```

```python
def cft(g, k1, k2):
    """Numerically evaluate the Fourier Transform of g for the given frequencies"""    
    result = np.zeros((len(k1), len(k2)), dtype=complex)
    
    # Loop over all frequencies and calculate integral value
    for i, kk1 in enumerate(k1):
        for l, kk2 in enumerate(k2):
            # Evaluate the Fourier Integral for a single frequency ff, 
            # assuming the function is time-limited to abs(t)<5
            result[i][l] = complex_quad(lambda x, y: g(x, y)*np.exp(-j*kk1*x)*np.exp(-j*kk2*y), -xmax, xmax, -ymax, ymax)
    return (1/(2*np.pi))*result

def complex_quad(func, x1, x2, y1, y2):
    """Return definite integral of complex-valued g from a to b, 
    using Simpson's rule"""
    # 2501: Amount of used samples for the trapezoidal rule
    t1 = np.linspace(x1, x2, 25) 
    t2 = np.linspace(y1, y2, 25) 
    dt = t2[1]-t2[0]
    
    ans = 0
    for tt1 in t1:
        for tt2 in t2:
            ans += func(tt1,tt2)*dt*dt
        
    return ans
    
```

```python
def one(x,y):
    return 1
complex_quad(one, -2, 2, -2, 2)
```

```python
g = lambda x, y: np.exp(-(x)**2-(y)**2)
xmax = 4
dx = 0.1
x=np.arange(-xmax,xmax,dx)
ymax = 4
dy = 0.1
y=np.arange(-ymax,ymax,dy)
```

```python
four= cft(g,x,y)
fig = plt.figure()
X, Y = np.meshgrid(x,y)
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, np.abs(four) , cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

```python

fig = plt.figure()
X, Y = np.meshgrid(x,y)
nonfour= g(X,Y)
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, np.abs(nonfour) , cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```
