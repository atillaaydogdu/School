```python
% matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import cm
from scipy.stats import norm

j = complex(0,1)
```

```python
scl = 7

func = lambda x, y: np.sinc(np.sqrt((x/scl)**2+(y/scl)**2))

x0 = -10*scl
dx = 0.05 * scl
x=np.arange(x0,-x0,dx)
y0 = -10*scl
dy = 0.05*scl
y=np.arange(y0,-y0,dy)
```

```python
X, Y = np.meshgrid(x,y)
nonfour = func(X,Y)
```

```python
fig = plt.figure()
plt.imshow(nonfour)
plt.colorbar()
plt.show()
```

```python
wx = np.fft.fftfreq(X[0].size)*2*np.pi/dx
scale = dx/((wx[1]-wx[0])*scl**2)
print(scale)
```

```python
fig = plt.figure()
g=np.fft.fft2(nonfour)
G = np.fft.fftshift( g )

plt.imshow(np.abs(G))
plt.xlim([200-400*scale/2,200+400*scale/2])
plt.ylim([200-400*scale/2,200+400*scale/2])
plt.colorbar()
plt.show()
```
