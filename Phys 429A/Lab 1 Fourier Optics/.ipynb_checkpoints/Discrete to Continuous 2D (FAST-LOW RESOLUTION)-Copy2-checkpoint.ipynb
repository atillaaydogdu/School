```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D 
from matplotlib import cm
from scipy.stats import norm

j = complex(0,1)
```

# Define the Aperture

```python
def func(x, y, radius = 1):
    X = x
    Y = y
    
    R = 0.1*np.sqrt(X**2+Y**2)
    return np.heaviside(R, 1) - np.heaviside(R-radius, -1)

x0 = -512
dx = 1
x=np.arange(x0,-x0,dx)
y0 = -512
dy = 1
y=np.arange(y0,-y0,dy)
```

# Hole Getting Smaller and Smaller...

```python
X, Y = np.meshgrid(x,y)
nonfour = func(X, Y, 4)

g=np.fft.fft2(nonfour)
G = np.fft.fftshift( g )
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,8))
axes[0].imshow(nonfour)
axes[1].imshow(np.abs(G))
plt.show()
```

```python
X, Y = np.meshgrid(x,y)
nonfour = func(X, Y, 2)

g=np.fft.fft2(nonfour)
G = np.fft.fftshift( g )
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,8))
axes[0].imshow(nonfour)
axes[1].imshow(np.abs(G))
plt.show()
```
