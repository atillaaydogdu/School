```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from uncertainties import ufloat
from uncertainties import unumpy
from scipy.optimize import curve_fit

import seaborn as sns
```

```python
paths = ['datafiles/head_angle_3/speed3.txt', 'datafiles/head_angle_3/speed4.txt', 'datafiles/head_angle_3/speed4a.txt',
         'datafiles/head_angle_3/speed5.txt', 'datafiles/head_angle_3/speed5a.txt', 'datafiles/head_angle_3/speed5b.txt',
        'datafiles/head_angle_3/speed6.txt','datafiles/head_angle_3/speed7.txt','datafiles/head_angle_3/speed8.txt',
        'datafiles/head_angle_3/speed9.txt','datafiles/head_angle_3/speed10.txt','datafiles/head_angle_3/speed11.txt',
        'datafiles/head_angle_3/speed13.txt']
```

```python
dfs = []
for filename in paths:
        df = pd.read_csv(filename, sep='\t', skiprows=5, index_col=0).astype(float) 
        dfs.append(df)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9,4))
[ax.hist(dfs[i]['Drag (N)'], bins=25, alpha=0.7, label='Speed {}'.format(i+1)) for i in range(13)]

ax.set_xlabel('Drag Force (N)')
ax.set_ylabel('Number of Measurements')
ax.grid()
ax.legend(loc=(1.02, 0.1))
fig.tight_layout()

plt.savefig('speed_var.png', dpi=300)
plt.show()
```
