```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from uncertainties import ufloat
from uncertainties import unumpy

%config InlineBackend.figure_format = 'svg'
```

```python
paths = ['datafiles/wing_test_1', 'datafiles/wing_test_2']
```

```python
codes_to_speeds = {'3':ufloat(3.57, 0.2), '4':ufloat(5.57, 0.2), '4a':ufloat(6.45, 0.2), '5':ufloat(8.3, 0.2),
                  '5a':ufloat(9.60, 0.2), '5b':ufloat(11.4, 0.2), '6':ufloat(14.9, 0.2), '7':ufloat(18.1, 0.2),
                   '8':ufloat(19.9, 0.2), '9':ufloat(21.7, 0.2), '10':ufloat(22.0, 0.2), '11':ufloat(22.9, 0.2),
                   '13':ufloat(23.5, 0.2)}                   
```

# Open up Data

Open up datafiles from each 10 second run and compute the mean and standard deviation for drag/lifting force. Each set of tests for a particular object (i.e testing a bunch of different wind speeds) is intended to be stored in a single DataFrame.

```python
def get_data(path):
    speed_codes = np.array([])
    lift_means = np.array([]); lift_stds = np.array([])
    drag_means = np.array([]); drag_stds = np.array([])
    
    for filename in os.listdir(path):
        df = pd.read_csv(path+'/'+filename, sep='\t', skiprows=5, index_col=0).astype(float) 
        speed_codes = np.append(speed_codes, filename[6:-4])
        lift_means = np.append(lift_means,  df['Lift (N)'].mean())
        lift_stds = np.append(lift_stds, df['Lift (N)'].std())
        drag_means = np.append(drag_means, df['Drag (N)'].mean())
        drag_stds = np.append(drag_stds, df['Drag (N)'].std())
        
    speeds = np.array([codes_to_speeds[speed_code] for speed_code in speed_codes])
    df =  pd.DataFrame({'speed_means': unumpy.nominal_values(speeds), 'speed_stds': unumpy.std_devs(speeds),
                         'lift_means':lift_means, 'lift_stds': lift_stds, 'drag_means':drag_means,
                         'drag_stds':drag_stds})
    
    return df.sort_values('speed_means')
```

Get the DataFrames.

```python
test1 = get_data(paths[0])
test2 = get_data(paths[1])
```

```python
display(test1.head())
```

# Plotting

```python
fig, ax = plt.subplots(1, 2, figsize=(10,3))

# Plot 1
ax[0].errorbar(test1['speed_means'], test1['lift_means'], xerr=test1['speed_stds'],
               yerr=test1['lift_stds'], ls='', color='k', alpha=0.4)
l1 = ax[0].scatter(test1['speed_means'], test1['lift_means'], s=15, label='Test 1')
ax[0].errorbar(test2['speed_means'], test2['lift_means'], xerr=test2['speed_stds'],
               yerr=test2['lift_stds'], ls='', color='k', alpha=0.4)
l2 = ax[0].scatter(test2['speed_means'], test2['lift_means'], s=15, label='Test 2')
ax[0].legend(handles=[l1,l2])

# Plot 2
ax[1].errorbar(test1['speed_means'], test1['drag_means'], xerr=test1['speed_stds'],
               yerr=test1['drag_stds'], ls='', color='k', alpha=0.4)
d1 = ax[1].scatter(test1['speed_means'], test1['drag_means'], s=15, label='Test 1')
ax[1].errorbar(test2['speed_means'], test2['drag_means'], xerr=test2['speed_stds'],
               yerr=test2['drag_stds'], ls='', color='k', alpha=0.4)
d2 = ax[1].scatter(test2['speed_means'], test2['drag_means'], s=15, label='Test 2')
ax[1].legend(handles=[d1,d2])

[a.set_xlabel('Wind Speed (m/s)') for a in ax.ravel()]
[a.set_ylabel('Force (N)') for a in ax.ravel()]
[a.grid() for a in ax.ravel()]

ax[0].set_title('Lifting Force')
ax[1].set_title('Drag Force')

plt.show()
```
