```python
import numpy as np
import matplotlib.pyplot as plt
```

# Question 5

```python
def G_5(f,K):
     return 1-np.sin((2*K+1)*np.pi*f)/((2*K+1)*np.sin(np.pi*f))
```

```python
f = np.linspace(-1/2, 1/2, 1000)

fig, ax = plt.subplots(1,1,figsize=(6,3))
ax.plot(f, G_5(f, 4)**2, label='K=4')
ax.plot(f, G_5(f, 16)**2, label='K=16')

ax.set_ylabel('$S_Y(f)$')
ax.set_xlabel('f')
ax.set_title('SDF for Filter Applied to $\sigma_\epsilon=1$ White Noise')
ax.grid()
ax.legend()

plt.savefig('A5_q5.png', dpi=300)
```

# Question 6

```python
def G_6_sq(f,K):
     return 4*np.sin(K*np.pi*f)**4/(K * np.sin(np.pi*f))**2
```

```python
f = np.linspace(-1/2, 1/2, 1000)

fig, ax = plt.subplots(1,1,figsize=(8,3))
ax.plot(f, G_6_sq(f, 1), label='K=1')
ax.plot(f, G_6_sq(f, 2), label='K=2')
ax.plot(f, G_6_sq(f, 4), label='K=4')
ax.plot(f, G_6_sq(f, 8), label='K=8')
ax.plot(f, G_6_sq(f, 16), label='K=16')

ax.axvspan(1/2, 1/4, alpha=0.2, color='blue')
ax.axvspan(1/(2*2), 1/(4*2), alpha=0.2, color='orange')
ax.axvspan(1/(2*4), 1/(4*4), alpha=0.2, color='green')
ax.axvspan(1/(2*8), 1/(4*8), alpha=0.2, color='red')
ax.axvspan(1/(2*16), 1/(4*16), alpha=0.2, color='purple')


ax.set_ylabel('$|G_K(f)|^2$')
ax.set_xlabel('f')
ax.set_title('Response Function to Filter')
ax.grid()
ax.legend()

plt.savefig('A5_q6.png', dpi=300)
```

# Question 7

```python
f = np.linspace(-1/2, 1/2, 1000)
SX = f**4
SZ = np.concatenate((SX[500:1000],SX[0:500]))
SZ_f = np.concatenate((np.zeros(400),SZ[400:600], np.zeros(400)))
SX_f = np.concatenate((SZ_f[500:1000],SZ_f[0:500]))

fig, ax = plt.subplots(1,4,figsize=(12,2.5))
ax[0].plot(f, SX)
ax[0].set_xlabel('$f$')
ax[0].set_ylabel('$S_X(f)$')

ax[1].plot(f, SZ)
ax[1].set_xlabel('$f$')
ax[1].set_ylabel('$S_Z(f)$')

ax[1].axvline(-0.1, color='k', ls='--')
ax[1].axvline(0.1, color='k', ls='--')

ax[2].plot(f, SZ_f)
ax[2].set_xlabel('$f$')
ax[2].set_ylabel('$S_Z(f)$ (Filtered)')

ax[3].plot(f, SX_f)
ax[3].set_xlabel('$f$')
ax[3].set_ylabel('$S_X(f)$ (Filtered)')


ax[0].grid()
ax[1].grid()
ax[2].grid()
ax[3].grid()

ax[0].set_title('Original SDF')
ax[1].set_title('SDF after (i)')
ax[2].set_title('SDF after (ii)')
ax[3].set_title('SDF after (iii)')

fig.tight_layout()
plt.savefig('a5_q7.png', dpi=300)
```
