```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.stattools
```

# Question 2

```python
def Fejer(f,N):
     return (1/N)*(np.sin(N*np.pi*f)/np.sin(np.pi*f))**2
```

```python
ff = np.linspace(-1/2, 1/2, 1000)
DD1 = 1-Fejer(ff,64)/64
DD2 = 1-Fejer(ff,128)/128
DD3 = 1-Fejer(ff,256)/256
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,3))
ax.plot(ff,DD1, label='N=64')
ax.plot(ff,DD2, label='N=128')

ax.set_xlabel('f')
ax.set_ylabel('$1-\mathcal{F}(f)/N$')

ax.grid()
ax.legend()

ax.set_title('Plot of $1-\mathcal{F}(f)/N$ for Different N Values')

plt.savefig('a6q2.png', dpi=300)
```

# Question 3

```python
ff = np.linspace(-np.pi/2, np.pi/2, 500)
DD = np.sin(ff)

acvs_DD_unbiased = statsmodels.tsa.stattools.acovf(DD, unbiased=True)
acvs_DD_biased = statsmodels.tsa.stattools.acovf(DD, unbiased=False)
```

```python
fig, ax = plt.subplots(2,1,figsize=(8,6))
ax[0].plot(ff,DD, color='k')

ax[0].set_xlabel('t')
ax[0].set_ylabel('Time Series Values [arb]')

ax[0].grid()

ax[0].set_title('Time Series (500 Points)')

ax[1].plot(acvs_DD_unbiased/acvs_DD_unbiased[0], label='Unbiased')
ax[1].plot(acvs_DD_biased/acvs_DD_biased[0], label='Biased')

ax[1].grid()
ax[1].set_xlabel(r'$\tau$')
ax[1].set_ylabel(r'$s_{\tau}/s_0$')
ax[1].legend()
ax[1].set_title('Unbiased vs. Biased Autocorrelation Sequence')

fig.tight_layout()
plt.savefig('a6q3.png', dpi=300)
```
