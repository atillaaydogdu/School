```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1/2, 1/2, 100)
K1 = 2
K2 = 5
K3 = 7
K4 = 9
```

```python
y1 = np.sin((2*K1+1)*np.pi*x)/np.sin(np.pi*x)
y2 = np.sin((2*K2+1)*np.pi*x)/np.sin(np.pi*x)
y3 = np.sin((2*K3+1)*np.pi*x)/np.sin(np.pi*x)
y4 = np.sin((2*K4+1)*np.pi*x)/np.sin(np.pi*x)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,3))
ax.plot(x,y1, label='K=2')
ax.plot(x,y2, label='K=5')
ax.plot(x,y3, label='K=7')
ax.plot(x,y4, label='K=9')


ax.set_xlabel('f')
ax.set_ylabel(r'$dS^{(I)}/df$')
ax.set_title(r'Plots of $dS^{(I)}/df$ for Different $K$ values', fontsize=20)

ax.legend()
ax.grid()

plt.savefig('A4q6.png', dpi=300)
```
