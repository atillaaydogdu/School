```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Omega(E):
    return 1-E**2

def d_ln(E):
    return (2*E)/(E**2-1)

def T(E):
    return (E**2-1)/(2*E)
```

```python
E = np.arange(-1,1,0.01)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10,3))

ax[0].plot(E, Omega(E))
ax[1].plot(E, d_ln(E))
ax[2].plot(E, T(E))

ax[0].set_xlabel('$E/E_{max}$')
ax[1].set_xlabel('$E/E_{max}$')
ax[2].set_xlabel('$E/E_{max}$')

ax[0].set_ylabel('$\propto \Omega(E)$')
ax[1].set_ylabel('$\propto \partial \ln\Omega(E)/\partial E$')
ax[2].set_ylabel('$\propto T(E)$')

ax[1].set_ylim(-10,10)
ax[2].set_ylim(-10,10)

fig.suptitle(r'Statistical Quantities For a Nuclear Spin System', fontsize=16)
fig.tight_layout()
fig.subplots_adjust(top=0.86)

plt.savefig('diff_stat_quan.png', dpi=300)
```
