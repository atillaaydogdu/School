# Plot the Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

Define the binomial distribution.

```python
def P(N, m, p):
    return (factorial(N)/(factorial((N+m)/2)*factorial((N-m)/2))
            *(p**((N+m)/2))*(1-p)**((N-m)/2))
```

## Case p=1/2

Define all possible values for each case.

```python
p=1/2
```

Create arrays of the possible steps that can be taken.

```python
m5 = np.arange(-5,6,2)
m10 = np.arange(-10,11,2)
m20 = np.arange(-20,21,2)
```

Define the Gaussian for each case.

```python
def norm(x, u, sig):
    return 1/(np.sqrt(2*np.pi)*sig)*np.exp(-((x-u)**2)/(2*sig**2))

x5 = np.arange(-5,5,0.01)
x10 = np.arange(-10,10,0.01)
x20 = np.arange(-20,20,0.01)

norm5 = norm(x5, 5*(2*p-1), np.sqrt(4*5*p*(1-p)))
norm10 = norm(x10, 10*(2*p-1), np.sqrt(4*10*p*(1-p)))
norm20 = norm(x20, 20*(2*p-1), np.sqrt(4*20*p*(1-p)))
```

### Plot

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4))

ax[0].bar(m5, P(5,m5,p))
ax[1].bar(m10, P(10,m10,p))
ax[2].bar(m20, P(20,m20,p), label='Binomial')

ax[0].plot(x5, norm5, c='r')
ax[1].plot(x10, norm10, c='r')
ax[2].plot(x20, norm20, c='r', label='Gaussian')

[a.grid() for a in ax]

ax[0].set_title('$N=5$')
ax[1].set_title('$N=10$')
ax[2].set_title('$N=20$')


ax[0].set_ylabel('Probability Density')
ax[1].set_xlabel('$m$')
ax[2].legend()


fig.suptitle('Binomial Distribution vs. Approximate Gaussian', fontsize=16)
fig.tight_layout()
fig.subplots_adjust(top=0.84)

plt.show()
```

## Case p=2/3

Redefine the value of $p$.

```python
p=2/3
```

Resolve for the Gaussian function.

```python
norm5 = norm(x5, 5*(2*p-1), np.sqrt(4*5*p*(1-p)))
norm10 = norm(x10, 10*(2*p-1), np.sqrt(4*10*p*(1-p)))
norm20 = norm(x20, 20*(2*p-1), np.sqrt(4*20*p*(1-p)))
```

### Plot

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4))

ax[0].bar(m5, P(5,m5,p))
ax[1].bar(m10, P(10,m10,p))
ax[2].bar(m20, P(20,m20,p), label='Binomial')

ax[0].plot(x5, norm5, c='r')
ax[1].plot(x10, norm10, c='r')
ax[2].plot(x20, norm20, c='r', label='Gaussian')

[a.grid() for a in ax]

ax[0].set_title('$N=5$')
ax[1].set_title('$N=10$')
ax[2].set_title('$N=20$')


ax[0].set_ylabel('Probability Density')
ax[1].set_xlabel('$m$')
ax[2].legend(loc='upper left')


fig.suptitle('Binomial Distribution vs. Approximate Gaussian', fontsize=16)
fig.tight_layout()
fig.subplots_adjust(top=0.84)

plt.show()
```

## A Peculiar Observation

It is worth noting that the Gaussian curve is only half as tall as the bar plots. This may intially seem like an error, but it is not, for the bar plots have a *gap* every second step where an observation cannot occur. The Gaussian function, remember, is a probability density function: it only tells us about the probability of an event in a specific interval. When the interval is large compared to the spacing of the gaps (large $N$) the Gaussian distribution as a probability density function does its job fine. For large $N$ then, one can image the blue bars as a "Gaussian function" that alternates between 0 and some positive value super rapidly. The average of such a function would be half as high; this is the red Gaussian curves plotted above.
