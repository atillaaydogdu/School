# Functions

```python
def img_shuffle(img):
    L = len(img[0])
    new_img = np.empty([L,L, 3], dtype=int)
    
    for i in range(L):
        for j in range(L):
            new_img[i][j] = img[(i+j)%L][(i+2*j)%L]
    return new_img
```

```python
def img_array(img, num_times):
    L = len(img[0])
    arr = np.empty([num_times, L, L, 3], dtype=int)
    for i in range(num_times):
        arr[i] = img
        img = img_shuffle(img)
    return arr
```
