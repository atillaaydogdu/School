## Functions

The function below takes in an image with pixel values $p(x_i, y_i)$ and creates a new image $p'(x_i, y_i)$ such that

$$p'(x_i, y_i) = p\left((x_i+y_i)\%L, (x_i+2y_i)\% L\right) $$

where $L$ is the length of the image.

```python
def img_shuffle(img):
    L = len(img[0])
    new_img = np.empty([L,L, 3], dtype=int)
    
    for i in range(L):
        for j in range(L):
            new_img[i][j] = img[(i+j)%L][(i+2*j)%L]
    return new_img
```

The function below takes in an image function and does the transformation

$$p \to p' \to p'' \to ... \to p^{(n)} $$

where $n$ is the "numtimes" argument in the function below.

```python
def img_array(img, num_times):
    L = len(img[0])
    arr = np.empty([num_times, L, L, 3], dtype=int)
    for i in range(num_times):
        arr[i] = img
        img = img_shuffle(img)
    return arr
```

# Find Number of Iterations Required

Create the transformations $p \to ...$ for each image and store them in arrays.

```python
arr_10 = img_array(img_10, 100)
arr_32 = img_array(img_32, 100)
arr_124 = img_array(img_124, 100)
arr_256 = img_array(img_256, 300)
```
