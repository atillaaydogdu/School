```python
def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap
```

# Show the Changing Histogram With Temperature

```python
upper_int = 100000
lower_int = 0
mode_std = 0.015
```

```python
cold_path = 'datafiles/mar_21_2019/gamma_data_cold.csv'
warm_path = 'datafiles/mar_21_2019/gamma_data_mid.csv'
```

```python
def get_df(path):
    
    df = pd.read_csv(path, sep=',', index_col=0).astype(float)
    df['ratio']=(df['Integral 1200']-df['Integral 0'])/(df['Integral 7400']-df['Integral 0'])
    df.index = pd.to_datetime(df.index,unit='s')-timedelta(hours=7)
        
    df = df[(df['ratio']<1) & (df['ratio']>0)] 
    df = df[(df['Integral 7400']<upper_int) & (df['Integral 7400']>lower_int)] 
    
    return df
```

```python
df_cold = get_df(cold_path)
df_warm = get_df(warm_path)
```

```python
[warm_mode, _, _], _ = Functions.mode_ratio(df_warm, mode_std)
[cold_mode, _, _], _ = Functions.mode_ratio(df_cold, mode_std)
```

```python
fig, ax = plt.subplots(figsize=(7,3))

df_cold['ratio'].hist(bins=500, ax=ax, alpha=0.5, label='8$^\circ$C')
df_warm['ratio'].hist(bins=500, ax=ax, alpha=0.5, color='r', label='20$^\circ$C')

ax.axvline(warm_mode, c='k', linestyle='-')
ax.axvline(cold_mode, c='k', linestyle='--')

ax.set_xlim(0.4,0.8)
#ax.set_title('Cold vs. Warm Photon Energy Deposition Speed')
ax.set_xlabel('E(1.2$\mu$s)/E(7.4$\mu$s)')
ax.set_ylabel('Number of Events')

ax.legend(prop={'size': 20})
fig.tight_layout()
plt.savefig('plots/report_plots/gamma_rat.png', dpi=300)
plt.show()
```

```python
[warm_mode_tote, _, _], _ = Functions.mode_energy(df_warm, mode_std=2000)
[cold_mode_tote, _, _], _ = Functions.mode_energy(df_cold, mode_std=2000)
```

```python
fig, ax = plt.subplots(figsize=(7,3))

df_cold['Integral 7400'].hist(bins=500, ax=ax, alpha=0.5, label='8$^\circ$C')
df_warm['Integral 7400'].hist(bins=500, ax=ax, alpha=0.5, color='r', label='20$^\circ$C')

ax.axvline(warm_mode_tote, c='k', linestyle='-')
ax.axvline(cold_mode_tote, c='k', linestyle='--')

ax.set_xlim(0,30000)
#ax.set_title('Cold vs. Warm Photon Total Energy Deposited')
ax.set_xlabel('E(7.4$\mu$s) [arb]')
ax.set_ylabel('Number of Events')

ax.legend(prop={'size': 20})
fig.tight_layout()
plt.savefig('plots/report_plots/gamma_e', dpi=300)
plt.show()
```
