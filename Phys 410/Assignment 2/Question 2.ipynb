# Question 2

The following plot contains the curves

$$r=A\left[\csc\left(\frac{\pi}{\alpha}\theta\right)\right]^{\alpha/\pi} $$

for $\alpha=\pi/3$ and for different values of $A$

```python
a = np.pi/3
```

The function shown above is defined.

```python
def r(t, A):
    return A*(np.sin(t*np.pi/a))**(-a/np.pi) 

theta = np.linspace(0.001, a-0.001, 1000)
```

Plots are created.

```python
fig = plt.figure(figsize=(7,7))

ax = plt.subplot(111, projection='polar')
ax.set_thetalim(0,a)
ax.set_rlim(0,6)

ax.plot(theta, r(theta, 1), label='A=1') 
ax.plot(theta, r(theta, 2), label='A=2') 
ax.plot(theta, r(theta, 3), label='A=3') 
ax.plot(theta, r(theta, 4), label='A=4') 

ax.set_title('Flow lines for Simple Fluid Flow in Wedge')
ax.legend(loc='best')

plt.show()
```
