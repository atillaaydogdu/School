# Code for Checking Complex Ellipse Question

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0,2*np.pi, 1000)
```

Consider the equation

$$z=ae^{i\phi}+be^{-i \phi} $$

where $a=a_r+ia_c$ and $b=b_r+ib_c$. We can also write 

$$a=|a|e^{is_1}$$  
$$b=|b|e^{is_2}$$  

# First Case

$$a=-1+i \hspace{5mm}  \text{and} \hspace{5mm} b=2+2i$$

Since $$\frac{s_1+s_2}{2} = \frac{\pi}{2}$$ the ellipse should be rotated by an angle of $\pi/2$.

```python
ar = -1
ac = 1
br = 2
bc = 2
```

## The Real Ellipse

```python
a = np.sqrt(ar**2+ac**2)
b= np.sqrt(br**+bc**2)


def x(t):
    return a*np.cos(t)+b*np.cos(t)

def y(t):
    return a*np.sin(t)-b*np.sin(t)


fig, ax = plt.subplots()
ax.scatter(x(time), y(time))

ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```

## The Rotated Ellipse

```python
def x(t):
    return ar*np.cos(t)-ac*np.sin(t)+br*np.cos(t)+bc*np.sin(t)

def y(t):
    return ar*np.sin(t)+ac*np.cos(t)-br*np.sin(t)+bc*np.cos(t)

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(x(time), y(time))
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```

# Case 2

$$a=3+3i \hspace{5mm}  \text{and} \hspace{5mm} b=2i$$

Since $$\frac{s_1+s_2}{2} = \frac{3\pi}{4}$$ the ellipse should be rotated by an angle of $3\pi/4$.

```python
ar = 3
ac = 3
br = 0
bc = 2
```

## The Real Ellipse

```python
a = np.sqrt(ar**2+ac**2)
b= np.sqrt(br**+bc**2)


def x(t):
    return a*np.cos(t+2)+b*np.cos(t+2)

def y(t):
    return a*np.sin(t+2)-b*np.sin(t+2)


fig, ax = plt.subplots()
ax.scatter(x(time), y(time))

ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```

## The Rotated Ellipse

```python
def x(t):
    return ar*np.cos(t)-ac*np.sin(t)+br*np.cos(t)+bc*np.sin(t)

def y(t):
    return ar*np.sin(t)+ac*np.cos(t)-br*np.sin(t)+bc*np.cos(t)

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(x(time), y(time))
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```
