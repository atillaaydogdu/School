# Code for Checking Complex Ellipse Question

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0,2*np.pi, 1000)
```

```python
ar = -1
ac = 1
br = 2
bc = 2

def x(t):
    return ar*np.cos(t)-ac*np.sin(t)+br*np.cos(t)+bc*np.sin(t)

def y(t):
    return ar*np.sin(t)+ac*np.cos(t)-br*np.sin(t)+bc*np.cos(t)

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(x(time), y(time))
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```

```python
ar = np.sqrt(ar**2+ac**2)
ac = 0
br = np.sqrt(br**+bc**2)
bc = 0

def x(t):
    return ar*np.cos(t+2)-ac*np.sin(t+2)+br*np.cos(t+2)+bc*np.sin(t+2)

def y(t):
    return ar*np.sin(t+2)+ac*np.cos(t+2)-br*np.sin(t+2)+bc*np.cos(t+2)


fig, ax = plt.subplots()
ax.scatter(x(time), y(time))

ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```
