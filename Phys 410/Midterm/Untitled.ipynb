```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(y, C):
    return np.log((y-C)/np.sin(C))+((y-C)/np.sin(C))*np.cos(C)

def get_xy(C):
    if C>0:
        y1 = np.linspace(C, 5, 1000)
        x1 = x(y1, C) 
    elif C<0:
        y1 = np.linspace(-5, C, 1000)
        x1 = x(y1, C) 
    
    return (x1, y1)
```

```python
C_array = np.arange(-2*np.pi/3, 2*np.pi/3, np.pi/6)
xy_array = [get_xy(C) for C in C_array]
```

```python
fig, ax= plt.subplots(figsize=(12, 8))

for (x, y) in xy_array:
    ax.plot(x, y)
    
ax.axhline(y=np.pi, xmin=0, xmax=0.4, linewidth=6, color='k', label='Capacitor')
ax.axhline(y=-np.pi, xmin=0, xmax=0.4, linewidth=6, color='k')
ax.set_xlim(-5, 5)

ax.legend()
ax.set_title('Equipotential Lines')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```
