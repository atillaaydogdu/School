```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import spherical_jn, spherical_yn, legendre
```

Define $v=2mV_0R^2/\hbar^2$

# Part B

Define function for computing $\delta_l$

```python
def delta_l(kR, v, l):
    return np.arctan( (kR*spherical_jn(n=l, z=kR, derivative=True)*spherical_jn(n=l, z= np.sqrt(kR**2+v))\
                    - np.sqrt(kR**2+v)*spherical_jn(n=l, z=kR)*spherical_jn(n=l, z= np.sqrt(kR**2+v), derivative=True))
                     / (kR*spherical_yn(n=l, z=kR, derivative=True)*spherical_jn(n=l, z= np.sqrt(kR**2+v))\
                    - np.sqrt(kR**2+v)*spherical_yn(n=l, z=kR)*spherical_jn(n=l, z= np.sqrt(kR**2+v), derivative=True)) )
```

Define $\sigma/R^2$ and $k^2 \sigma$

```python
def cross_tot_R2(kR, v, order=100):
    return 4*np.pi/(kR**2) * np.sum([(2*l+1)*np.sin(delta_l(kR, v, l))**2 for l in range(order)], axis=0)

def cross_tot_k2(kR, v, order=100):
    return 4*np.pi * np.sum([(2*l+1)*np.sin(delta_l(kR, v, l))**2 for l in range(order)], axis=0)
```

Compute values for a few different values of $V_0$

```python
kRs = np.linspace(1,16,100)
vs = np.array([4,8,16,24,32])
cross_div_R2s = [cross_tot_R2(kRs, v) for v in vs]
cross_times_k2s = [cross_tot_k2(kRs, v) for v in vs]
```

Plot

```python
[plt.plot(kRs, cross_div_R2, label='$2mV_0R^2/\hbar^2$={}'.format(v)) for (cross_div_R2,v) in zip(cross_div_R2s, vs)]
plt.legend()
plt.xlabel('$kR$')
plt.ylabel('$\sigma/R^2$ [Cross Section]')
plt.title('Cross section vs. $kR$')
plt.show()
```

In general, as the energy of the incident particle (or the radius of the potential) increases, the ratio $\sigma/R^2$ decreases. In addition, there are a few resonant energies that result in local maxima of the cross section. **This corresponds to the fixed potential plot in part C of the previous question, only now we are in 3D**.

# Part C

Get array of different values for $V_0$.

```python
v = np.linspace(0,250,10000)
```

Obtain the total cross section.

```python
cross_tot_R2_V0_dep = cross_tot_R2(0.02, v, order=50)
```

Plot.

```python
plt.plot(v,cross_tot_R2_V0_dep)
plt.xlabel('$2mV_0R^2/\hbar^2$ [Depth of Well]')
plt.ylabel('$\sigma/R^2$ [Cross Section]')
plt.title('$kR$=0.02 ')
plt.grid()
plt.show()
```

Note that we get the same result if we just use the zeroth order term ($\delta_0$ as we found on paper).

```python
cross_tot_R2_V0_dep = cross_tot_R2(0.02, v, order=1)

plt.plot(v,cross_tot_R2_V0_dep)
plt.xlabel('$2mV_0R^2/\hbar^2$ [Depth of Well]')
plt.ylabel('$\sigma/R^2$ [Cross Section]')
plt.title('$kR$=0.02 ')
plt.grid()
plt.show()
```

We can find the peaks using scipy's signal libraries.

```python
from scipy.signal import find_peaks
peaks, _ = find_peaks(cross_tot_R2_V0_dep, height=15000)
```

The locations of the peaks are at the following values of $2mV_0R^2/\hbar^2$ for $kR=0.02$.

```python
v[peaks]
```

The physical interpretation of these peaks is as follows:

Given an incindent free particle with wave function $e^{ikx}$, and a spherical well with a given radius $R$, there are certain potentials $-V_0$ that will result in a very large scattering cross section, but since these peaks are narrow, it requires a very particular value of $V_0$.

**This corresponds to the fixed energy plot in part C of the previous question, only now we are in 3D**.
